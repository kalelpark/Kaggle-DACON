```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Ellipse
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'

from math import pi
import warnings
warnings.filterwarnings("ignore")

other_color = "salmon"
korea_color = "skyblue"

block_time_card = pd.read_csv('data/블록별 시간대별소액결제건수 카드소비패턴.csv', encoding = 'cp949')
block_time_card.rename(columns = {'서울시민업종코드(UPJONG_CD)' : 'upjoing_cd', '기준년월(YM)' : 'ym',
                                  '시간대구간(TIME)' : 'time', '고객주소블록코드(BLOCK_CD)' : 'block_cd',
                                  '소액결제건수(MICRO_PYM)' : 'micro_pym', '카드이용금액계(AMT_CORR)' : 'amt_corr'}, inplace = True)

card_use_pattern = pd.read_csv('data/블록별 성별연령대별 카드소비패턴.csv', encoding='cp949')
card_use_pattern.rename(columns={'서울시민업종코드(UPJONG_CD)' : 'upjoing_cd', '기준년월(YM)' : 'ym' , 
                         '고객주소블록코드(BLOCK_CD)' : 'block_cd', '성별(GEDNER)' : 'gender',
                        '연령대별(AGE)' : 'age', '카드이용금액계(AMT_CORR)' : 'amt_corr',
                         '카드이용건수계(USECT_CORR)' : 'usect_corr'}, inplace = True)
card_use_pattern['Sex'] = card_use_pattern['gender'] == 'M'

card_use_code = pd.read_csv('data/카드소비 업종코드.csv', encoding='cp949')
card_use_code.rename(columns={'업종코드(UPJONG_CD)' : 'upjoing_cd', '대분류(CLASS1)' : 'class1' , 
                         '중분류(CLASS2)' : 'class2', '소분류(CLASS3)' : 'class3'}, inplace = True)

agg_card_where = pd.read_csv('data/집계구별 일별소비지역별 카드소비패턴.csv' , encoding='cp949')
agg_card_where.rename(columns= {'가맹점주소광역시도(SIDO)' : 'sido', '가맹점주소시군구(SGG)' : 'sgg', 
                                '업종대분류(UPJONG_CLASS1)' : 'class1', '기준일자(YMD)' : 'ymd',
                               '고객주소집계구별(TOT_REG_CD)' : 'tot_reg_cd', '카드이용금액계(AMT_CORR)' : 'amt_corr',
                               '카드이용건수계(USECT_CORR)' : 'usect_crr'}, inplace = True)
agg_card_where.drop('sgg', axis = 1, inplace = True)

agg_card_time = pd.read_csv('data/집계구별 일별시간대별 카드소비패턴.csv', encoding = 'cp949')
agg_card_time.rename(columns={'업종대분류(UPJONG_CLASS1)': 'class1', '기준일자(YMD)' : 'ymd',
                              '시간대구간(TIME)' : 'time', '고객주소집계구별(TOT_REG_CD)' : 'tot_reg_cd',
                             '카드이용금액계(AMT_CORR)' : 'amt_corr', '카드이용건수계(USECT_CORR)' : 'usect_corr'}, inplace = True)
```

# 카드 사용자 고객의 나이 분포

- 데이터의 분포를 봤을 때, 카드 사용자 고객의 나이 분포는 20 ~ 40대가 가장 많으며, 그 다음으로는 50 ~ 70대입니다.<br>
  제 추리상, 카드가 어떤 카드인지는 명확히 모르지만, 10대들에 10세 이상을 기준으로 한다면,<br> 
  아마 체크카드에 관한 dataset이지 않을까 싶습니다.

```python
fig, ax = plt.subplots(1, 1, figsize = (20, 9))
temp = card_use_pattern['age'].value_counts().sort_index()
ax.bar(temp.index, temp, width = 0.55, color='skyblue',
       linewidth=0.7)

for i in temp.index:
    ax.annotate(f"{temp[i]}", 
                   xy=(i, temp[i] + 2),
                   va = 'center', ha='center',fontweight='light',
                   color='#4a4a4a')
    
for s in ['top', 'left', 'right']:
    ax.spines[s].set_visible(False)
fig.text(0.1, 0.95, '카드 사용자 고객의 나이 분포', fontsize=30, fontweight='bold')
plt.show()
```

# 카드 사용자 고객의 성별/나이 분포

- 데이터셋을 봤을 때, 성별과 나이의 데이터 분포가 정확히 동일하다는 것을 알 수 있습니다.

```python
card_use_pattern['gender'] = card_use_pattern['gender'].apply(lambda x : 'ETC' if x not in ['F', 'M'] else x)
temp = card_use_pattern[card_use_pattern['gender'] != 'ETC'].groupby(['gender'])['age'].value_counts().unstack().sort_index()
temp.rename(index={'F' : 'Female', 'M' : 'Male'}, inplace = True)
man = temp.loc['Male']
woman = -temp.loc['Male']

fig, ax = plt.subplots(1, 1, figsize = (16, 8))
ax.bar(man.index, man, width = 0.4, color = 'salmon', alpha = 0.5, label = 'Male')
ax.bar(woman.index, woman, width = 0.4, color = 'skyblue', alpha = 0.5, label = 'Female')
plt.rcParams.update({'font.size' : 20})
for i in man.index:
    ax.annotate(f"{man[i]}",
               xy = (i, man[i] + 3),
               va = 'center', ha = 'center', fontweight = 'light', fontfamily = 'serif', fontsize = 13)

    
for i in woman.index:
    ax.annotate(f"{-woman[i]}",
               xy = (i, woman[i] - 3),
               va = 'center', ha = 'center', fontweight = 'light', fontfamily = 'serif', fontsize = 13)

for s in ['top', 'left', 'right', 'bottom']:
    ax.spines[s].set_visible(False)
    
fig.text(0.16, 0.95, '카드 사용자 고객의 성별/나이 분포', fontsize=30, fontweight='bold')
ax.set_yticks([])
plt.show()
```

# 성별에 따른 카드 이용 건수 및 이용 금액

- 남성의 경우보다 여성의 경우가 카드 이용 건수가 조금 더 많다는 것을 알 수 있습니다. <br> 
  하지만, 남성의 경우가 여성의 경우보다 카드 이용금액은 더욱 많다는 것을 알 수 있습니다.

```python
x = card_use_pattern.groupby(by='gender').sum().index
x = x.values
x[0] = 'Female'
x[1] = 'Male'

y1 = card_use_pattern.groupby(by='gender').sum()['usect_corr']
y = card_use_pattern.groupby(by='gender').sum()['amt_corr']

plt.figure(figsize = (20, 10))
plt.subplot(1, 2, 1)
color = sns.color_palette('pastel')[:18]
plt.title('성별에 따른 카드 이용 건수 (%)')
plt.xlabel('성별')
plt.ylabel('카드이용 건수')
plt.pie(y1, labels= x, colors= color, explode = [0.05, 0.05],  autopct='%1.1f%%', textprops= {'fontsize' : 20})

plt.subplot(1, 2, 2)
plt.title('성별에 따른 카드 이용 금액 계 (%)')
plt.xlabel('성별')
plt.ylabel('카드 이용 금액 계')
plt.pie(y, labels= x, colors= color, explode = [0.05, 0.05],  autopct='%1.1f%%', textprops= {'fontsize' : 20})
plt.suptitle('성별에 따른 카드 이용 건수 및 금액 계 (%)', fontsize = 25)
plt.show()
```

# 카드 이용 내역

- 카드 사용자들의 카드 이용 내역을 봤을 때, 요식/유흥을 카드로 많이 이용하고, (먹는건 못 참죠!) <br>
  그 다음 순으로는, 유통, 의료에 카드 이용을 많이 한다는 것을 알 수 있습니다.

```python
temp = card_use_code['upjoing_cd']
temp2 = card_use_code['class1']
card_joing = dict()
for i in range(len(temp)):
    temp[i] = temp[i].replace('s', 'S')
    card_joing[temp[i]] = temp2[i]
card_use_code['upjoing_cd'] = temp
card_use_pattern['class'] = card_use_pattern['upjoing_cd'].map(card_joing)

x = card_use_pattern['class'].value_counts().index
y = card_use_pattern['class'].value_counts().values
plt.rcParams.update({'font.size' : 15})
plt.figure(figsize = (20, 10))
plt.barh(x,y, color = 'skyblue', alpha = 0.5)
plt.title('업종별 카드 이용 현황', fontsize = 30)
plt.xlabel('카드 이용 현황', fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylabel('업종별')
plt.show()
```

# 업종별 카드 이용 금액 현황

- 카드 사용에는 유통에 가장 많은 돈을 소비하게 되며, 그 다음 요식/유흥에 가장 많이 카드로 소비하는 금액이 많다는 것을 알 수 있습니다.

```python
temp = card_use_pattern.groupby(by = 'class').sum()
temp
x = temp.index
y = temp['amt_corr']
plt.figure(figsize = (20, 10))
plt.rcParams.update({'font.size' : 15})
plt.figure(figsize = (20, 10))
plt.barh(x,y, color = 'salmon')
plt.title('업종별 카드 이용 금액 현황(단위 : 1e7)', fontsize = 30)
plt.xlabel('카드 금액 현황', fontsize = 20)
# plt.xlim(826681, 100000000)
plt.xticks(np.arange(800000, 100000000, 5000000))
plt.show()
```

# 지역에 따른 업종별 카드 이용 현황

- 지역에 따라 업종별 카드 이용 내역을 알아봤습니다. 도시별로 카드 이용 내역을 분류해서 봤을 때, <br>
  지역이 서울인 데이터가 상당수를 차지하고, 지역이 서울이 아닌 데이터가 분석을 하기에는 다소, 데이터의 양이 적다고 판단했기에, 묶었습니다. <br>
  (하단 경기/인천/부산은 따로 분석)
  
<br>

- 지역별로 카드를 사용하는 용도는 비슷하다는 것을 알 수 있으며, 역시 먹는 건 못참는다는 것을 알 수 있습니다.

```python
agg_card_where['country'] = agg_card_where['sido'] == '서울'
temp = agg_card_where.groupby('country')['class1'].value_counts().to_frame()
temp.columns = ['amount']
temp = temp.reset_index(drop = False)
temp = temp.pivot(index = 'country', columns='class1')['amount']
temp['all'] = temp.sum(axis = 1)

for c in temp.columns:
    temp[c] = temp[c] / temp['all']*100
temp = temp.drop(['all'], axis = 1)
temp = temp.T
temp.columns = ['서울 외', '서울']
labels = temp.index.values
y1 = temp['서울 외'].values
y2 = temp['서울'].values
explode = [0.02 for i in range(14)]
# print(len(labels))
plt.figure(figsize = (20, 10))
plt.subplot(1, 2, 1)
color = sns.color_palette('pastel')[:18]
plt.pie(y1, labels = labels, colors=color, explode = explode, autopct='%1.1f%%', textprops= {'fontsize' : 10})
plt.title('서울 외 업종별 카드 이용 현황')
plt.subplot(1, 2, 2)
plt.pie(y2, labels = labels, colors=color, explode = explode, autopct='%1.1f%%',  textprops= {'fontsize' : 10})
plt.title('서울 업종별 카드 이용 현황')
plt.suptitle('지역에 따른 업종별 카드 이용 현황', fontsize = 25)
plt.show()
```

# 경기/강원/인천/부산 업종별 카드 이용 현황
